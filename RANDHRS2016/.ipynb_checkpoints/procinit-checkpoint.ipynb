```python
import pandas as pd
import numpy as np

def prune_and_quantize(df: pd.DataFrame, m: int = 2, k: int = 10, method: str = "quantile") -> pd.DataFrame:
    """
    Return a copy of df where:
      1) Columns with < m unique (non-NaN) values are dropped.
      2) Numeric columns with > k unique (non-NaN) values are quantized to k levels.
         - method='quantile': equal-frequency bins via qcut; values replaced by bin medians.
         - method='uniform': equal-width bins; values replaced by bin centers.
      3) NaNs are preserved; non-numeric columns are unchanged (aside from step 1).
    """
    out = df.copy()

    # 1) Drop low-variability columns
    keep = [c for c in out.columns if out[c].nunique(dropna=True) >= m]
    out = out[keep]

    # 2) Quantize numeric columns
    num_cols = out.select_dtypes(include=[np.number]).columns
    for c in num_cols:
        s = out[c]
        # count unique excluding NaN
        uniq = s.dropna().unique()
        if len(uniq) > k and k > 0:
            print(c)
            if method == "quantile":
                # qcut can drop bins if there are many ties; handle that
                try:
                    bins = pd.qcut(s, q=k, duplicates="drop")
                except ValueError:
                    # if all values identical or too few distinct, skip
                    continue
                # replace each bin by its median value (quantization representative)
                med = s.groupby(bins).transform("median")
                out[c] = med
            elif method == "uniform":
                # equal-width bins on min..max, replace with bin centers
                v = s.values
                mask = ~np.isnan(v)
                if mask.sum() == 0:
                    continue
                lo, hi = np.nanmin(v), np.nanmax(v)
                if lo == hi:
                    continue
                # create k bins; np.digitize returns 1..k
                edges = np.linspace(lo, hi, num=k+1)
                centers = (edges[:-1] + edges[1:]) / 2.0
                idx = np.digitize(v[mask], edges[1:-1], right=False)
                qv = np.full_like(v, np.nan, dtype=float)
                qv[mask] = centers[idx]
                out[c] = qv
            else:
                raise ValueError("method must be 'quantile' or 'uniform'")

    return out
```

```python
df2=prune_and_quantize(df1,k=26)
```
